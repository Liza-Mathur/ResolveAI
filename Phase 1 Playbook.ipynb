Let's take a look at data first. 

```python
import pandas as pd
import json
```

```python
products = pd.read_json('data/products.json')
```

```python
with open('data/products.json') as products:
    
    products = json.load(products)
```

```python
products
```

```python
products_data_text = []
for p in products:
    text = """ 
    Product name = p['name'], Category = p['category'], productid = p['product_id'], description = p['description'], materials = p['materials'], sizes = p['sizes'], colors = p['colors'], use_case = p['use_case'], gender=p['gender'], instock = p['in_stock'], rating = p['rating'] 
    """
    products_data_text.append({"text" : text, "id":p['product_id']})
```

```python
len(products_data_text)
```

```python
with open('data/reviews.json') as reviews:
    reviews = json.load(reviews)
reviews_text_data = []
for r in reviews:
    text = """
    product_id = r['product_id'], customer_id = p['customer_id'], rating5=r['rating5'], title=r['title'], review=r['review']
    """
    reviews_text_data.append({'text':text, 'id':r['review_id']})
len(reviews_text_data)
```

```python
with open("data/faqs.md", "r", encoding="utf-8") as f:
    text = f.read()
faqs = {"text": text, "id": "faqs"}
```

```python
with open("data/loyalty_programme.md","r", encoding="utf-8") as l:
    text = l.read()
loyalty_programme = {"text":text, "id":"loyalty program"}
```

```python
with open("data/returns_policy.md","r", encoding="utf-8") as rp:
    text = rp.read()
return_policy = {"text":text, "id":"Return Policy"}
```

```python
with open("data/shipping_policy.md","r", encoding="utf-8") as sp:
    text = sp.read()
shipping_policy = {"text":text, "id":"Shipping Policy"}
```

```python
with open("data/warranty_policy.md","r", encoding="utf-8") as wp:
    text = wp.read()
warranty_policy = {"text":text, "id":"Warranty Policy"}
```

```python
text_data_all = products_data_text + reviews_text_data + [faqs] + [loyalty_programme] + [return_policy] + [shipping_policy] + [warranty_policy]
```
